# Statistical insights

## Insights
- Valor de Geração: desvio padrão alto, o que indica uma grande variabilidade na geração de energia.
- Valor de Geração Limitada: a média de geração limitada (86.36 MW) é consideravelmente menor que a de geração total  (82.13MW), o que sugere que em média, há um curtailment significativo nas usinas eólicas. O alto descio padrão também aponta que em algumas situações o curtailment pode ser bem expressivo, enquando em outras não.
- Valor de Disponibilidade: A média de disponibilidade de energia é superior à média de geração e de geração limitada e seu desvio padrão mostra que há grandes variações de disponibilidade entre parques.
- Valor de Geração de Referência: A geração de referência, que pode ser uma medida de produção esperada ou ideal, tem uma média mais baixa do que a disponibilidade, mas ainda assim mostra uma variação considerável. Isso pode indicar que a geração real frequentemente não atinge os níveis ideais, possivelmente devido a limitações operacionais.
- Valor de Geração de Referência Final: A geração de referência final tem a menor média entre todos os valores de geração. Isso pode indicar que, após a aplicação das restrições (como o curtailment), a geração final de energia eólica fica bem abaixo da geração ideal ou até da geração limitada.
- Valor do Vento Verificado: Sua média de velocidade e desvio padrão indicam que as condições do vento variam muito.
- Flag Dado Vento Inválido: O valor médio de 0.21 indica que cerca de 21% dos dados de vento podem ser inválidos, o que pode ser um indicativo de que as condições de vento nem sempre são confiáveis ou bem medidas. O desvio padrão elevado sugere que a qualidade dos dados de vento pode variar muito entre os momentos de medição.
- Valor de Geração Estimada: O desvio padrão é bem alto, podendo indicar alta variabilidade das condições do vento, capacidade das usinas ou falhas no modelo de previsão. 
- Valor de Geração Verificada: A média de geração verificada é ligeiramente superior à estimada (13.34 MW vs. 12.30 MW). O desvio padrão é novamente alto, o que indica que a geração verificada pode variar bastante ao longo do tempo.

- Os dados altamente assimétricos (como o vento verificado) podem exigir atenção especial, pois outliers podem representar condições críticas que influenciam significativamente a produção de energia.

# z-score e o valor-p para identificar outliers na coluna val_geracao

Ajuda a detectar valores atípicos na geração de energia eólica para que se possam entender melhor os dados e tomar decisões informadas sobre a operação.

## Insights
- Neste teste a quantidade de outliers foi baixa, em torno de 5%
- Os valores de média e desvio padrão indicam que a geração de energia eólica tem uma distribuição com grande variação 
- Os outliers têm uma média muito mais alta (362.87 MW) do que a média geral (82.13 MW)
- O desvio padrão dos outliers (100.61 MW) também é relativamente alto, o que indica que esses outliers não são uniformes, mas estão distribuídos de forma ampla

### Possíveis causas dos outliers
- Eventos exepcionais de ventos
- Erro de medição ou falha nos sensores

```python
# Calculating the mean and standard deviation of the 'val_geracao' column
mean = df_const_off['val_geracao'].mean()
std = df_const_off['val_geracao'].std()
# Calculating the z-score for the 'val_geracao' column
df_const_off['z_score'] = (df_const_off['val_geracao'] - mean) / std
# Calculating the p-value for the z-score
df_const_off['p_value'] = 2 * (1 - stats.norm.cdf(abs(df_const_off['z_score'])))
# Calculating the threshold for the p-value
threshold = 0.05
# Filtering the dataframe for outliers
outliers = df_const_off[df_const_off['p_value'] < threshold]
# Displaying the outliers
display(outliers)
# Displaying the number of outliers
print("Number of outliers: ", len(outliers))
# Displaying the percentage of outliers
print("Percentage of outliers: ", len(outliers) / len(df_const_off) * 100)
# Displaying the mean and standard deviation of the 'val_geracao' column
print("Mean: ", mean)
print("Standard deviation: ", std)
# Displaying the mean and standard deviation of the 'val_geracao' column for the outliers
print("Mean of outliers: ", outliers['val_geracao'].mean())
print("Standard deviation of outliers: ", outliers['val_geracao'].std())
```
